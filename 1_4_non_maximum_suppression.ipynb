# Non-maximum Suppression

```python
import numpy as np
from src.utils import round_matrix
```

```python
prefix = "assets/corner_response/step4_"
left_corner_response = np.load(prefix + "left_corner_response.npy")
right_corner_response = np.load(prefix + "right_corner_response.npy")

expected_nms_left_response = np.load(prefix + "left_nms_corner_response.npy")
expected_nms_right_response = np.load(prefix + "right_nms_corner_response.npy")
```

```python
expected_nms_left_response
```

```python
def non_maximum_suppression(corner_response, image_width, image_height):
    ext_corner_response = np.pad(
        corner_response, pad_width=1, mode="edge"
    )

    nms = []
    kernel_size = 3
    for i in range(image_height):
        for j in range(image_width):
            if corner_response[i, j] == 0:
                continue

            neighborhood = ext_corner_response[i : i + kernel_size, j : j + kernel_size]
            if corner_response[i, j] == np.max(neighborhood):
                nms.append((i, j))

    return nms
```

```python
image_width = 210
image_height = 200

nms_left_response = non_maximum_suppression(left_corner_response, image_width, image_height)
nms_right_response = non_maximum_suppression(right_corner_response, image_width, image_height)
```

```python
print(np.array_equal(nms_left_response, expected_nms_left_response))
print(np.array_equal(nms_right_response, expected_nms_right_response))
```
