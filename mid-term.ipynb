## Question 1
You are given two sets of 2D points, `A` and `B`, representing corresponding feature coordinates in two images. Write a Python function called `procrustes_align(A, B)` that:

- Centers both sets of points.
- Computes the best-fitting rotation matrix using SVD of the cross-covariance matrix.
- Applies the rotation to align `A` to `B`.

Note 1: Assume the numpy Python package has been imported before the execution of the function `procrustes_align(A, B)`. Please see the example below!

Note 2: Your implemented function should return the aligned points and the rotation matrix.

Note 3: Do not round any result in your implementation.

````python
def procrustes_align(A, B):
    A = np.asarray(A)
    B = np.asarray(B)
    
    centroid_A = np.mean(A, axis=0)
    centroid_B = np.mean(B, axis=0)
    
    A_centered = A - centroid_A
    B_centered = B - centroid_B
    
    H = A_centered.T @ B_centered
    
    U, S, Vt = np.linalg.svd(H)
    
    d = np.linalg.det(U @ Vt)
    D = np.diag([1, d])
    R = U @ D @ Vt
    
    A_centered_aligned = A_centered @ R.T
    
    A_aligned = A_centered_aligned + centroid_B
    
    return A_aligned, R
````
