## A. Implement compute_homo

```python
def compute_homo(matched_corner_pairs):
    A = np.zeros((2 * len(matched_corner_pairs), 9))
    
    for i, (pt1, pt2) in enumerate(matched_corner_pairs):
        x1, y1 = pt1
        x2, y2 = pt2
        
        A[2*i] = [0, 0, 0, x1, y1, 1, -y2*x1, -y2*y1, -y2]
        A[2*i + 1] = [x1, y1, 1, 0, 0, 0, -x2*x1, -x2*y1, -x2]
        
    print("Matrix A:")
    print(A)
    
    _U,_D,Vt = np.linalg.svd(A) 
    h = Vt[-1]
    H = h.reshape(3, 3)
    return H
```

```python
def find_intersection_of_two_3D_lines(p1, p2, p3, p4):
    # Direction vectors for both lines
    v1 = p2 - p1
    v2 = p4 - p3
    
    # Vector connecting the starting points
    c = p3 - p1
    
    # Cross product of the direction vectors
    v1_cross_v2 = np.cross(v1, v2)
    
    # Squared magnitude of the cross product
    norm_sq = np.dot(v1_cross_v2, v1_cross_v2)
    
    # Guard against parallel or collinear lines (where cross product is zero vector)
    if np.isclose(norm_sq, 0):
        raise ValueError("Lines are parallel or collinear; no unique intersection point.")
    
    # Calculate the scalar t1 for the first line
    c_cross_v2 = np.cross(c, v2)
    t1 = np.dot(c_cross_v2, v1_cross_v2) / norm_sq
    
    # Substitute t1 back into the first line equation to find the exact intersection
    exact_intersection = p1 + t1 * v1
    
    # Round to the nearest integer and return
    return np.round(exact_intersection)

p1 = np.array([5, 1, 10])
p2 = np.array([6, 1, 7])
p3 = np.array([8, -4, 8])
p4 = np.array([3, -1, 9])

X, Y, Z = find_intersection_of_two_3D_lines(p1, p2, p3, p4)
print(f"The intersection point of two 3D lines: X = {X}, Y = {Y}, Z = {Z}")
```
