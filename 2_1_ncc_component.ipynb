```python
import numpy as np
from PIL import Image
```

```python
def compute_NCC_component(image, corner_list, window_size):
    margin = window_size // 2
    
    ncc_components = []
    for y, x in corner_list:

        # Indexes for the window around the corner
        y_l = y-margin
        y_r = y+margin
        x_l = x-margin
        x_r = x+margin
        
        # Skip if the window outside the image boundary
        if (y_l < 0 or y_r >= image.shape[0] or
            x_l < 0 or x_r >= image.shape[1]):
            continue
        
        window = image[y_l : y_r + 1,
                       x_l : x_r + 1]

        mean = np.mean(window)        
        zero_mean = window - mean
        zero_mean_squared_sum = np.sum(zero_mean ** 2)
        
        # ignore zero denominator
        if zero_mean_squared_sum == 0:
            continue

        denominator = np.sqrt(zero_mean_squared_sum)
        ncc_component = zero_mean / denominator
        
        ncc_components.append((y, x, ncc_component))
    
    return ncc_components
```

```python
left_image = np.array(Image.open("assets/original_left_image.png").convert('L'))
right_image = np.array(Image.open("assets/original_right_image.png").convert('L'))

prefix = "assets/corner_response/step4_"
left_corner_list = np.load(prefix + "left_nms_corner_response.npy")
right_corner_list = np.load(prefix + "right_nms_corner_response.npy")

window_size = 15
NCCs_left_component = compute_NCC_component(left_image, left_corner_list, window_size)
NCCs_right_component = compute_NCC_component(right_image, right_corner_list, window_size)

```
