# Optimal Homography Transformation

```python
import numpy as np
```

```python
P_1_old = np.array([
    [2450, -4550, 100],
    [212.5, -487.5, -5112.5],
    [-0.7, -0.7, -0.1]])
P_1_new = np.array([
    [425, -1260, 5055],
    [2768.75, -4182.5, -958.75],
    [-0.65, -0.74, 0.17]])

P_2_old = np.array([
    [2450, -4550, -100],
    [712.5, -337.5, -5225],
    [-0.7, -0.7, -0.2]])
P_2_new = np.array([
    [425, -1260, 5055],
    [2768.75, -4182.5, -958.75],
    [-0.65, -0.74, 0.17]])
```

```python
np.round(np.linalg.inv(P_1_old), 2)
```

```python
np.round(P_1_new @ np.linalg.inv(P_1_old), 2)
```

```python
np.round(P_2_new @ np.linalg.inv(P_2_old), 2)
```
